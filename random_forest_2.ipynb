```python
df["SubType"] = ""
```

```python
def cal_subtype(row):
    GM = (row["Q1"] + row["Q6"] + row["Q11"])/3
    MG = (row["Q2"] + row["Q7"] + row["Q12"])/3
    ET = (row["Q3"] + row["Q8"] + row["Q13"])/3
    PW = (row["Q4"] + row["Q9"] + row["Q14"])/3
    CI = (row["Q5"] + row["Q10"] + row["Q15"])/3
    res=[]
    if GM > 3:
        res.append("GM")
    if MG > 3:
        res.append("MG")
    if ET > 3:
        res.append("ET")    
    if PW > 3:
        res.append("PW") 
    if CI > 3:
        res.append("CI") 
    if len(res)==0:
        res.append("Not")
#     print("GM", GM)
#     print("MG", MG)
#     print("ET", ET)
#     print("PW", PW)
#     print("CI", CI)
    return "-".join(res)
```

```python
cal_subtype(df.iloc[2])
```

```python
for i, row in df.iterrows():
    df.loc[i,"SubType"] = cal_subtype(row)
```

```python
df
```

```python
df["SubType"].value_counts()
```

```python
y=df["SubType"]
X=df.drop(columns=["SubType",'Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10', 'Q11',
       'Q12', 'Q13', 'Q14', 'Q15'])
```

```python
X.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test)
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
print(classification_report(y_test, predictions))
```

```python
pd.DataFrame({"features":X_train.columns,"importances":rf_model.feature_importances_}).sort_values(by="importances",ascending=False)
```
