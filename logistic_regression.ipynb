```python
import pandas as pd
```

```python
file = ("cleaned_data_numeric.csv")
data_df = pd.read_csv(file)
data_df
```

```python
# Create bins to convert total scores to boolean values (Yes=conspiracy theorist, No=not conspiracy theorist)
size_bins = [0, 3.5, 5]
group_names = ["No", "Yes"]
```

```python
# Categorize score based on the bins
data_df["Outcome"] = pd.cut(data_df["TotalScore"], size_bins, labels=group_names)

data_df.head()
```

```python
# We dont need the TotalScore column anymore so i dropped it
data_df = data_df.drop(["TotalScore"], axis=1)
```

```python
# assigning dependent and independent variables 
y = data_df["Outcome"]
X = data_df.drop(columns="Outcome")
```

```python
# splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
   y, random_state=1, stratify=y)
```

```python
# checking distribution 
X_train.shape
```

```python
# create model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
   max_iter=200,
   random_state=1)
# what are these parameters? 
```

```python
classifier.fit(X_train, y_train)
```

```python
# create predictions
y_pred = classifier.predict(X_test)
```

```python
# view predicitions 
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# test accuracy 
from sklearn.metrics import accuracy_score
print(f'model is {(accuracy_score(y_test, y_pred)*100).round(2)}% accurate')
```
