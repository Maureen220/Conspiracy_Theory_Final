```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy import create_engine
import psycopg2
import config
from config import db_password
```

# Loading Table 1 Into Database

```python
#load in first dataset
government_malfeasance_df = pd.read_csv('Resources/government_malfeasance.csv')
government_malfeasance_df
```

```python
government_malfeasance_df['government_malfeasance']=government_malfeasance_df['government_malfeasance'].round(0).astype(int)
government_malfeasance_df
```

```python
# # Connect to sql database 
# # If you were loading from Jupyter Notebook to SQL, this works. 
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/mock_database_final_project"
engine = create_engine(db_string)
government_malfeasance_df.to_sql(name='government_malfeasance', con=engine, if_exists = 'replace')
```

```python
#import table from sql database 
government_malfeasance_DB_df = pd.read_sql_table("government_malfeasance", con=engine, index_col="index")
```

```python
government_malfeasance_DB_df
```
